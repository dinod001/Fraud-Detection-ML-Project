### 01. Import Dependecies

```python
import pandas as pd
```

### 02. Load Data

```python
df = pd.read_csv("data/processed/feature_engineered.csv")
df.head()
```

### 03. Feature Binning

```python
df['age'].describe()
```

```python
def binning_age(age):
    if age >= 18 and age <= 25:
        return "Young"
    elif age >= 26 and age <= 35:
        return "Adult"
    elif age >= 36 and age <= 55:
        return "Middle-aged"
    elif age >= 56:
        return "Senior"
    else:
        return "Unknown"
```

```python
df['age'] = df['age'].map(binning_age)
```

### 04. Saving Data

```python
df.to_csv("data/processed/feature_binning.csv",index=False)
```
